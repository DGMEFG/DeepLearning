**1. 激活函数**

```python
x = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.relu(x)

plot(x.detach(), y.detach(), 'x', 'relu(x)', figsize=(5, 2.5))
```

```python
# dy / dx * [1, 1, ..., 1]
y.backward(torch.ones_like(x), retain_graph=True)
plot(x.detach(), x.grad, 'x', 'grad of relu', figsize=(5, 2.5))
```

```python
y = torch.sigmoid(x)
plot(x.detach(), y.detach(), 'x', 'sigmod(x)', figsize=(5, 2.5))
```

```python
x.grad.zero_()
y.backward(torch.ones_like(x), retain_graph=True)
# 保留计算图是因为，需要 x.grad 不清空
plot(x.detach(), x.grad, 'x', 'grad of sigmoid', figsize=(5, 2.5))
```

```python
y = torch.tanh(x)
plot(x.detach(), y.detach(), 'x', 'tanh(x)', figsize=(5, 2.5))
```

```python
x.grad.zero_()
y.backward(torch.ones_like(x))
plot(x.detach(), x.grad, 'x', 'grad of tanh(x)', figsize=(5, 2.5))
```
